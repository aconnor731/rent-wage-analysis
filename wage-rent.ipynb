```python
import pandas as pd
```

```python
df = pd.read_csv('./State_MedianRentalPrice_AllHomes.csv')
```

```python
states = df['RegionName']
```

```python
ten = df.filter(regex='2010|RegionName', axis=1)
eleven = df.filter(regex='2011|RegionName', axis=1)
twelve = df.filter(regex='2012|RegionName', axis=1)
thirteen = df.filter(regex='2013|RegionName', axis=1)
fourteen = df.filter(regex='2014|RegionName', axis=1)
fifteen = df.filter(regex='2015|RegionName', axis=1)
sixteen = df.filter(regex='2016|RegionName', axis=1)
seventeen = df.filter(regex='2017|RegionName', axis=1)
eighteen = df.filter(regex='2018|RegionName', axis=1)
nineteen = df.filter(regex='2019|RegionName', axis=1)

ten_mean = ten.mean(1)
eleven_mean = eleven.mean(1)
twelve_mean = twelve.mean(1)
thirteen_mean = thirteen.mean(1)
fourteen_mean = fourteen.mean(1)
fifteen_mean = fifteen.mean(1)
sixteen_mean = sixteen.mean(1)
seventeen_mean = seventeen.mean(1)
eighteen_mean = eighteen.mean(1)
nineteen_mean = nineteen.mean(1)

all_years_means = []
all_years_means.append(pd.concat([states, ten_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, eleven_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, twelve_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, thirteen_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, fourteen_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, fifteen_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, sixteen_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, seventeen_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, eighteen_mean], axis=1, join_axes=[states.index]))
all_years_means.append(pd.concat([states, nineteen_mean], axis=1, join_axes=[states.index]))
```

```python
year = 2010
i = 0
while year < 2020:
    df = all_years_means[i]
    df.columns = ['States', str(year)]
    year = year + 1
    i = i + 1
```

```python
meanOfMedianRentsByYear_df = states
year = 2010
for item in all_years_means:
    meanOfMedianRentsByYear_df = pd.concat([meanOfMedianRentsByYear_df,
                                            item[str(year)]], axis=1,
                                           join_axes=[meanOfMedianRentsByYear_df.index])
    year = year + 1
```
